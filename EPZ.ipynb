```python

import seaborn as sns
sns.set_style("darkgrid")
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [6,4]
plt.rcParams["figure.autolayout"] = True
df=pd.read_excel(r"D:\EPZ Research\all parameter+predicted.xlsx",'SETTLEMENT')
def addlabels(x,y,z):
    for i in range(len(x)):
        plt.text(i+z,y[i],y[i],ha = 'center', rotation='vertical',color='black',fontsize=12,fontweight="bold")
X=df.iloc[0,1:6]
X_axis = np.arange(len(X))
Y=[]
for i in range(5):
    Y_Float=df.iloc[1,i+1]
    Y_int=int(Y_Float)
    Y.append(Y_int)
plt.bar(X_axis,Y,0.4,label = 'Settlement',color = 'red',edgecolor = 'black')
plt.xticks(X_axis, X)
plt.xlabel("Years",style='italic',color='blue',fontname="Times New Roman",fontweight="bold",fontsize=12)
plt.ylabel("Number of Industrial Building in EPZ Area",style='italic',color='blue',fontname="Times New Roman",fontweight="bold",fontsize=12))
addlabels(X, Y,0)
plt.ylim(0,160)
plt.savefig(r"D:\EPZ Research\EPZ Area.png",dpi=2000)
```
